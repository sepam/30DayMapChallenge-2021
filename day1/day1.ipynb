```python
import geopandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = geopandas.read_file('airports.csv')
df['geometry'] = geopandas.points_from_xy(df.longitude_deg, df.latitude_deg)
```

```python
df.head()
```

```python
dfs = df[['name', 'latitude_deg', 'longitude_deg']]
dfs.to_csv('airports-small.csv', index=False)
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
ax.set_aspect('equal')  # ensures aspect ratio is maintained
df.plot(ax=ax, markersize=1, alpha=0.5)
```
