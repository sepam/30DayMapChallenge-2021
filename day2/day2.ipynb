```python
import geopandas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = geopandas.read_file('hotosm_idn_jakarta_roads.gpkg')
```

```python
df.head()
```

```python
roads = df[df.highway.isin(['primary', 'secondary', 'tertiary'])]

mono_font = {'fontname':'monospace'}  # use monospace font for text on the map

fig, ax = plt.subplots(figsize=(20, 20))
ax.set_aspect('equal')  # ensures aspect ratio is maintained

ax.set_xticks([]) # removes the coordinate ticks on the map
ax.set_yticks([])

plt.title('Roads of Jakarta', fontsize=40, **mono_font)
plt.text(0.87, 0.02, 'Martijn Wieriks, 2021', transform=ax.transAxes, **mono_font)
roads.plot(ax=ax, color='#262323', linestyle='dotted', linewidth=1)
plt.savefig('map.png')  # save the file
```
